```python
import sys
import time
import random

def slow_print(text, delay=0.02):
    for c in text:
        print(c, end="", flush=True)
        time.sleep(delay)
    print()

def get_choice(prompt, valid_choices):
    while True:
        choice = input(prompt).lower()
        if choice in valid_choices:
            return choice
        slow_print("Invalid choice — try again!")

def restart():
    choice = input("\nWould you like to restart? (y/n): ").lower()
    if choice == "y":
        game()
    else:
        slow_print("\nFarewell, traveler of Skyrim.")
        sys.exit()

def death(reason):
    slow_print("\n⚔️ YOU DIED ⚔️")
    slow_print(f"CCause of death: {reason}")
    restart()

def game():
    slow_print("""
===============================================
        SKYRIM: THE BLAZE OF WHITERUN
===============================================

The land of Skyrim stands on the edge of ruin.
Rumors whisper of a terrible dragon, Mirmulnir,
descending from the sky to burn the villages
surrounding Whiterun. Jarl Balgruuf has sent out
a desperate call for any adventurer brave enough
to investigate and put an end to the beast.

You answer that call...

Your journey begins now.
===============================================
""")

#Step 1
    slow_print("YYou stand at the crossroads outside Whiterun.")
    slow_print("""
Three paths lie before you:
  1. The Western Road, leading into a dense forest rumored to be haunted.
  2. The Northern Pass, a snowy ravine crawling with bandits.
  3. The Eastern Plains, open fields patrolled by Whiterun guards.
""")

    step1 = get_choice("Choose your path (1/2/3): ", ["1", "2", "3"])

#Step 2
    if step1 == "1":
        slow_print("""
The forest seems to close around you, its shadows thickening.
A chill runs down your spine, and somewhere far off,
something howls. As you press deeper, a pair of glowing
eyes blink to life in the brush...
""")
        slow_print("""
What will you do?:
  A. Draw your weapon and stand your ground.
  B. Try to sneak around the creature.
""")
        step2 = get_choice("Choose A/B: ", ["a", "b"])

        if step2 == "a":
            slow_print("""
You face the beast - a massive frost wolf.
With a roar, it leaps! You manage to fend it off,
driving it away after a brutal struggle.
Wounded but alive, you press onward.
""")
        else:
            death("A frost wolf stalked you silently and tore you apart.")

    elif step1 == "2":
        slow_print("""
Snow crunches under your boots as the mountains loom overhead.
The wind howls like an angry spirit. Ahead, a group of Forsworn bandits
sit around a fire, sharpening their blades.
""")
        slow_print("""
What will you do?:
  A. Attempt to sneak past the bandits.
  B. Challenge them head-on.
""")
        step2 = get_choice("Choose A/B: ", ["a", "b"])

        if step2 == "b":
            death("The bandits overwhelmed you and slashed you to bits.")
        else:
            slow_print("""
You slip behind the rocks, silently crossing the ridge.
One bandit glances your way but shrugs it off. You escape
without a fight — your journey continues.
""")

    elif step1 == "3":
        slow_print("""
You walk across the rolling plains. The golden grass sways
in the breeze and a patrol of Whiterun guards spots you.
They greet you and warn of dangers ahead — dragon sightings.
""")
        slow_print("""
What will you do?:
  A. Ask to accompany the guards for protection.
  B. Continue alone, confident in your skills.
""")
        step2 = get_choice("Choose A/B: ", ["a", "b"])

        if step2 == "a":
            slow_print("""
The guards walk with you until the crossroads. Their company
bolsters your courage, and they share supplies before parting ways.
""")
        else:
            slow_print("""
You continue alone. The wind carries the faint smell of smoke.
Something is burning in the distance...
""")
    else:
        slow_print("Invalid choice — try again.")
        return game()

#Step 3 (barrow)
    barrow = random.choice([True, False])

    if barrow:
        slow_print("""
You arrive at an old Nordic barrow entrance. The stones hum faintly,
and ancient carvings glow with a ghostly light.
""")
        slow_print("""
What will you do?:
  A. Enter the barrow — dragons often roost near ruins.
  B. Make camp outside and rest.
""")
        step3b = get_choice("Choose A/B: ", ["a", "b"])

        if step3b == "b":
            death("A necromancer emerged from the barrow at night and ended your rest permanently.")
        else:
            slow_print("""
Inside, your footsteps echo. Cold air coils around you.
Faint whispers drift from deeper within the ruin...
""")

#Step 4 (barrow)
        slow_print("""
You come across a stone chamber. In the center lies a glowing
Word Wall — ancient power vibrating through the air.
""")
        slow_print("""
What will you do?:
  A. Approach the Word Wall.
  B. Ignore it and proceed carefully.
""")
        step4b = get_choice("Choose A/B: ", ["a", "b"])

        if step4b == "a":
            slow_print("""
The wall pulses with light as ancient knowledge floods into you.
You learn the first word of a Shout - 'Fus', the power of force.
""")
            has_shout = True
        else:
            slow_print("You move on, wary of ancient magic.")
            has_shout = False

#Step 5 (barrow)
        slow_print("""
You exit the barrow and climb a rocky ridge overlooking the plains.
Smoke rises from the direction of the Western Watchtower.
""")
        slow_print("""
What will you do?:
  A. Rush toward the smoke.
  B. Approach cautiously from the cliffs.
""")
        step5b = get_choice("Choose A/B: ", ["a", "b"])

        if step5b == "a":
            slow_print("You sprint toward the tower. A roar shakes the sky...")
        else:
            slow_print("You creep along the rocks, staying hidden from above.")

#Step 3 (giants)
    else:
        slow_print("""
As you continue down the path, the ground trembles beneath you.
A distant roar rolls across the plains - but not from any dragon.
Ahead, a group of giants herd mammoths across your route.
""")
        slow_print("""
What will you do?:
  A. Keep your distance and circle around the giants.
  B. Attempt to cross their path quickly.
""")
        step3g = get_choice("Choose A/B: ", ["a", "b"])

        if step3g == "b":
            death("A giant spotted you and smashed your skull in.")
        else:
            slow_print("""
You circle wide around the massive creatures, careful not to disturb
them. The earth still trembles from their steps as you move on.
""")

#Step 4 (giants)
        slow_print("""
The air grows warm as you reach a narrow canyon. Strange red crystals
jut from the rock walls, humming with energy. A smith from Whiterun
once claimed these were fragments of a fallen star.
""")
        slow_print("""
What will you do?:
  A. Inspect the glowing crystals.
  B. Ignore them and keep moving.
""")
        step4g = get_choice("Choose A/B: ", ["a", "b"])

        if step4g == "a":
            slow_print("""
You reach out - the crystal pulses violently. A shockwave erupts,
sending you flying. When you stand, you feel... different. Stronger.
You learn the first word of a Shout - 'Fus', the power of force.
""")
            has_shout = True
        else:
            slow_print("You press on, wary of strange magic in the canyon.")
            has_shout = False

#Step 5 (giants)
        slow_print("""
You climb out of the canyon and see smoke spiraling from the Western
Watchtower. A thunderous roar splits the air - unmistakably a dragon.
""")
        slow_print("""
What will you do?:
  A. Rush toward the tower.
  B. Approach cautiously from the hills.
""")
        step5g = get_choice("Choose A/B: ", ["a", "b"])

        if step5g == "a":
            slow_print("You race toward the smoke as the ground shakes...")
        else:
            slow_print("You move along the hillside, keeping low and quiet.")

#Step 6 (paths reconverge)
    slow_print("""
As you near the tower, Mirmulnir - the dragon - swoops overhead,
its wings beating like thunder. Flames engulf the watchtower.
""")
    slow_print("""
What will you do?:
  A. Join the Whiterun guards in battle.
  B. Try to strike from a hidden vantage.
""")
    step6 = get_choice("Choose A/B: ", ["a", "b"])

    if step6 == "b":
        death("The dragon spotted you easily and reduced you to ash.")
    else:
        slow_print("You charge into battle alongside the guards!")

#Step 7
    slow_print("""
The dragon lands, crushing a guard beneath its claws.
Its massive head turns toward you, smoke curling from its jaws.
""")
    slow_print("""
Two choices:
  A. Aim for its wings to limit its movement.
  B. Charge directly for its head.
""")
    step7 = get_choice("Choose A/B: ", ["a", "b"])

    if step7 == "b":
        death("You ran straight into the dragon's jaws. A bold but fatal move.")
    else:
        slow_print("Your attack wounds the dragon, grounding it.")

#Step 8
    slow_print("""
The beast roars in fury, sweeping its tail across the battlefield.
""")
    slow_print("""
What will you do?:
  A. Dodge the tail sweep.
  B. Try to block the attack.
""")
    step8 = get_choice("Choose A/B: ", ["a", "b"])

    if step8 == "b":
        death("The dragon's tail shattered your bones like twigs.")
    else:
        slow_print("You dodge, rolling to safety.")

#Step 9
    slow_print("""
The dragon staggers, one wing crippled. Flames rise in its throat.
This is your chance.
""")
    slow_print("""
What will you do?:
  A. Use your Shout (if you learned it).
  B. Strike with your weapon.
""")
    step9 = get_choice("Choose A/B: ", ["a", "b"])

    if step9 == "a":
        if has_shout:
            slow_print("""
You shout: 'FUS!'
The blast of force staggers the dragon, exposing its underbelly.
""")
        else:
            death("You tried to Shout... but had never learned one. The dragon incinerated you.")
    else:
        slow_print("You charge, weapon ready.")

#Step 10
    slow_print("""
With the dragon off balance, you have one final chance to slay it.
""")
    slow_print("""
Two choices:
  A. Leap onto its neck and strike.
  B. Charge at it's head on foot.
""")
    step10 = get_choice("Choose A/B: ", ["a", "b"])

    if step10 == "b":
        death("The dragon lunged forward and swallowed you whole.")
    else:
        slow_print("""
===============================================
          ⚔️  MIRMULNIR IS SLAIN! ⚔️
===============================================

Your blade pierces the dragon’s skull. Its roar fades into a
haunting whisper as its soul swirls around you, flames dancing
in the air until -

- it sinks into your very being.

The guards stand in stunned silence.

"You... you're Dragonborn," one whispers.

Whiterun is safe. For now.
===============================================
""")
        slow_print("Thank you for playing!")
        restart()

game()
```
